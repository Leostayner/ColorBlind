No item 2, para produzir cores reduzidas incluídas no
a gama de cores do monitor, o domínio de cores da paleta inicial é ligeiramente atenuado, escalonando das grandezas fotométricas relativas.

```python
result_protanopes   = sample_image.copy()
result_deuteranopes = sample_image.copy()

for i in range(3):
    result_protanopes[:, :, i]   = 0.992052 * result_protanopes[:, :, i] + 0.003974
    result_deuteranopes[:, :, i] = 0.957237 * result_deuteranopes[:, :, i] + 0.0213814
```

A partir disto, podemos converter os valores de RGB para LMS, multiplicando a matriz RGB (3,1) por uma matriz criada pelos pesquisadores, composta por constantes.

O LMS... **falar do LMS** (tem no artigo).

```python
#Funções
def matrix_multiply(matriz_P, matriz_D, imagem_P, imagem_D, size):
    for i in range(size[0]):
        for j in range(size[1]):
            matriz_Auxiliar  = np.array([[imagem_P[i][j][0]],[imagem_P[i][j][1]],[imagem_P[i][j][2]]])
            matriz_Auxiliar2 = np.array([[imagem_D[i][j][0]],[imagem_D[i][j][1]],[imagem_D[i][j][2]]])
        
            matriz_Auxiliar  = np.matmul(matriz_P, matriz_Auxiliar)
            matriz_Auxiliar2 = np.matmul(matriz_D, matriz_Auxiliar2)
        
            for k in range(3):    
                imagem_P[i][j][k] = matriz_Auxiliar[k][0]
                imagem_D[i][j][k] = matriz_Auxiliar2[k][0]
    
    return imagem_P, imagem_D
```

```python
result_protanopes_LMS = result_protanopes.copy() 
result_deuteranopes_LMS = result_deuteranopes.copy()

convertion_RGB2LMS = np.array([[17.8824, 43.5161, 4.11935],[3.45565, 27.1554, 3.86714],[0.0299566, 0.184309, 1.46709]])

for i in range(img_size[0]):
    for j in range(img_size[1]):
        x = result_deuteranopes_LMS[i][j].ravel()
        #np.swapaxes(x,0,1)
        result_deuteranopes_LMS[i][j] = np.dot(x, convertion_RGB2LMS)
        
#result_protanopes_LMS, result_deuteranopes_LMS = matrix_multiply(convertion_RGB2LMS, convertion_RGB2LMS, result_protanopes_LMS, result_deuteranopes_LMS, img_size)
              
fig = plt.figure(figsize=(15,15))

plt.subplot(1, 2, 1)
plt. title('Protanopes')
plt.axis('off')
#plt.imshow(result_protanopes_LMS)

plt.subplot(1, 2, 2)
plt.title('Deuteranopes')
plt.axis('off')
plt.imshow(result_deuteranopes_LMS)
```

```python
result_protanopes_LMS2 = result_protanopes.copy() 
result_deuteranopes_LMS2 = result_deuteranopes.copy()

convertion_LMS2LMS_P = np.array([[0, 2.02344, -2.52581],[0, 1, 0],[0, 0, 1]])
convertion_LMS2LMS_D = np.array([[1, 0, 0],[0.494207, 0, 1.24827],[0, 0, 1]])

result_protanopes_LMS2, result_deuteranopes_LMS2 = matrix_multiply(convertion_LMS2LMS_P, convertion_LMS2LMS_D, result_protanopes_LMS2, result_deuteranopes_LMS2, img_size)

fig = plt.figure(figsize=(15,15))

plt.subplot(1, 2, 1)
plt.title('Protanopes')
plt.axis('off')
plt.imshow(result_protanopes_LMS2)

plt.subplot(1, 2, 2)
plt.title('Deuteranopes')
plt.axis('off')
plt.imshow(result_deuteranopes_LMS2)
```
